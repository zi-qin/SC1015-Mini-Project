# Data Visualization and Analysis

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
pd.options.mode.chained_assignment = None  # default='warn'
```

**Numerical Data**

```python
# Import clean data
clean_df = pd.read_csv('clean_data.csv')

#Extract every single numerical variable
YrsCurEmp = pd.DataFrame(clean_df['YrsCurEmp'])
NumChild = pd.DataFrame(clean_df['NumChild'])
Stress = pd.DataFrame(clean_df['Stress'])
Age  = pd.DataFrame(clean_df['Age'])
Promo = pd.DataFrame(clean_df['Promo'])
OpProc = pd.DataFrame(clean_df['OpProc'])
Supvr = pd.DataFrame(clean_df['Supvr'])
IntLvEmp = pd.DataFrame(clean_df['IntLvEmp'])
```

```python
# Set up matplotlib figure with seven subplots
f, axes = plt.subplots(4, 2, figsize=(24, 24))

# Plot the basic uni-variate figures for numerical variable 
sb.histplot(data = YrsCurEmp, ax = axes[0,0])
sb.histplot(data = NumChild, ax = axes[0,1])
sb.histplot(data = Stress, ax = axes[1,0])
sb.histplot(data = Age, ax = axes[1,1])
sb.histplot(data = Promo, ax = axes[2,0])
sb.histplot(data = OpProc, ax = axes[2,1])
sb.histplot(data = Supvr, ax = axes[3,0])
sb.histplot(data = IntLvEmp, ax = axes[3,1])
```

```python
print(YrsCurEmp.describe())
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = YrsCurEmp, orient = "h")
```

## Correlations between numerical data variables

```python
# Extract only the numeric data variables
numDF = pd.DataFrame(clean_df[['YrsCurEmp', 'NumChild', 'Stress', 'Age', 'Promo', 'OpProc', 'Supvr', 'IntLvEmp']])

# Calculate the complete  correlation matrix
numDF.corr()
```

```python
# Heatmap of the Correlation Matrix
f = plt.figure(figsize=(12, 12))
sb.heatmap(numDF.corr(), vmin = -1, vmax = 1, annot = True, fmt = ".2f")
```

### Observations:

**YrsCurEmp** and **Age** have the highest correlation.  
**Stress** and **Supvr** have the lowest correlation.

*using **IntLvEmp** as a response variable,*

**Stress** has the highest correlation with **IntLvEmp**.
The higher the work stress levels of the employee, the higher the intention of leaving for employee.

**NumChild** has the lowest correlation with **IntLvEmp**.
Therefore, the higher the number of children in household, the lower the intention of leaving for employee. 

Therefore, stress levels (**Stress**) and Satisfaction for supervision (**Supvr**) can be the most influential factor or variable to predict the intentions of leaving for employees compared to other variables.
