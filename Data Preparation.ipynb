```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Importing data
df = pd.read_csv('data.csv')
df
```

### Selecting relevant variables 

The dataset consists of responses from survey participants to 508 questions which reflects their job experience and their intentions to change employer.

24 out of the 508 variables will be used in creating models to predict employee's intention to change employer. 

### Description of the dataset from the study:

Accessed from: https://www.icpsr.umich.edu/web/ICPSR/studies/26782/variables

> **Q3** : Years with current employer 

> **Q17** : Have thought about looking for a job with a different employer in the past year (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q44** : Current career choice was significantly influenced in a positive way by a high school teacher (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q45** : Current career choice was significantly influenced in a positive way by a high school guidance counselor (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q46** : Current career choice was significantly influenced in a positive way by my parents (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q47** : Current career choice was significantly influenced in a positive way by my brother or sister (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q48** : Current career choice was significantly influenced in a positive way by my spouse or partner (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q49** : Current career choice was significantly influenced in a positive way by my children (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q50** : Current career choice was significantly influenced in a positive way by a friend (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q76** : Number of children in household

> **Q84** : Breakdown of household care and maintenance (0-no spouse or partner, 1-all my partner's responsibility, 2-mostly my partner's responsibility, 3-responsibility shared equally, 4-mostly my responsibility, 5-all my responsibility)

> **Q121** : Gender discrimination is a significant barrier to advancement in my career field (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q123** : Discrimination on the basis of ethnic background is a significant barrier to advancement my career field (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q140** : Have to miss family activities due to the amount of time I must spend on work responsibilities (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q200** : Feel that he/she is being paid a fair amount for the work done (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q205** : Benefits received are as good as most other organizations offer. (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **Q212** : Enjoy coworkers (1-strongly disagree, 2-moderately disagree, 3-slightly disagree, 4-slightly agree, 5-moderately agree, 6-strongly agree)

> **DEMO4** : Education level (0-UNKNOWN CODE, 2-high school graduate or ged, 3-trade or technical training, 4-some college, no degree, 5-associate degree, 6-bachelor's degree, 7-master's degree, 8-professional degree, 9-doctorate)

> **WRKSTRS4** : Work stress (Overall Score using mean of 4 indicators)

> **AGE** : Age

> **JSSPROMO** : Satisfaction with promotion (undeclared range of scores, but higher scores denote higher satisfaction)

> **JSSOP** : Satisfaction with operating procedures (undeclared range of scores, but higher scores denote higher satisfaction)

> **JSSSUPER** : Satisfaction with supervision (undeclared range of scores, but higher scores denote higher satisfaction)

> **ILEMP5** : Intention to leave employer (overall score using mean of 5 indicators)


```python
# Extracting relevant variables from the dataset 
extracted_df=df[['Q3', 'Q17', 'Q44', 'Q45', 'Q46', 'Q47', 'Q48', 'Q49', 'Q50', 'Q76', 'Q84', 'Q121', 'Q123', 'Q140', 'Q200', 'Q205', 'Q212', 'DEMO4', 'WRKSTRS4', 'AGE', 'JSSPROMO', 'JSSOP', 'JSSSUPER', 'ILEMP5']]
extracted_df
```

```python
# Setting values which are 0 (unknown code) for 'DEMO4'
s=extracted_df.shape[0]
c=0;
while c<s:
    if extracted_df.at[c, 'DEMO4']==0:
        extracted_df.at[c, 'DEMO4']=np.nan
    c+=1
```

##### Removing & replacing null values 
- null values in 'Q17' column which will be used as a predictor variable for the model will be dropped to avoid 
misrepresentation of the data 
- filling the remaining null values with either median (numerical) or mode (categorical)

```python
# dropping rows with null values in 'Q17' column 
new_df=extracted_df.dropna(subset=['Q17'])
new_df
```

```python
mode=new_df.mode()
```

```python
# replacing null with median for numerical data
array=['Q3', 'Q76', 'WRKSTRS4', 'AGE', 'JSSPROMO', 'JSSOP', 'JSSSUPER', 'ILEMP5']
for n in array:
    d=new_df[n].dropna()
    m=np.median(d)
    new_df[n].fillna(m, inplace = True)
```

```python
# replacing null with mode for categorical data
array=['Q44', 'Q45', 'Q46', 'Q47', 'Q48', 'Q49', 'Q50', 'Q84', 'Q121', 'Q123', 'Q140', 'Q200', 'Q205', 'Q212', 'DEMO4']
for n in array:
    m=mode.at[0,str(n)]
    new_df[str(n)].fillna(m, inplace = True)
```

```python
new_df.info()
```

###### Creating a new variable 
- As our interest is in whether the presence of influence in career choice will affect an employee's intention to leave his/her employer, we will be converting responses for Q44-Q50 to either agree or disagree. If an employee agrees (option 4-6) to any of these questons, it would be considered as agree. Otherwise (option 1-3), it will be considered as disagree.

```python
# creating the new variable
inf_cols=new_df[['Q44', 'Q45', 'Q46', 'Q47', 'Q48', 'Q49', 'Q50']]
def custom(row):
    if row.sum()>7*3:
        return "Agree"
    return "Disagree"
inf_col = pd.DataFrame(inf_cols.apply(custom, axis=1))

# concatenated df with new column
n_df=pd.concat([new_df, inf_col], axis=1)
n_df
```

```python
# dropping the old variables
n_df.drop(['Q44', 'Q45', 'Q46', 'Q47', 'Q48', 'Q49', 'Q50'], axis=1, inplace=True)
n_df
```

##### Converting variables with float data type to integer data type 
- removing the redundant decimal points for all variables, other than 'WRKSTRS4' variable, 'ILEMP5' and newly generated variable

```python
# removing decimal points
float_cols=n_df.drop(['WRKSTRS4', 0, 'ILEMP5'], axis=1)
int_cols = pd.DataFrame(float_cols.applymap(int))
clean_df=pd.concat([int_cols, n_df[['WRKSTRS4', 0, 'ILEMP5']]], axis=1)
clean_df
```

##### Renaming categories in categorical variables 

```python
col1 = clean_df['Q84']

col1.replace(0, "No Partner", inplace=True)
col1.replace(1, "All Partner Responsibility", inplace=True)
col1.replace(2, "Mostly Partner Responsibility", inplace=True)
col1.replace(3, "Equal Responsibility", inplace=True)
col1.replace(4, "Mostly My Responsibility", inplace=True)
col1.replace(5, "All My Responsibility", inplace=True)

col2 = clean_df['DEMO4']

col2.replace(2, "High School Graduate or GED", inplace=True)
col2.replace(3, "Technical Training", inplace=True)
col2.replace(4, "Some college (No Degree)", inplace=True)
col2.replace(5, "Associate Degree", inplace=True)
col2.replace(6, "Bachelor Degree", inplace=True)
col2.replace(7, "Master Degree", inplace=True)
col2.replace(8, "Professional Degree", inplace=True)
col2.replace(9, "Doctorate", inplace=True)

col3 = clean_df['Q17']

col3.replace(1, "Disagree", inplace=True)
col3.replace(2, "Disagree", inplace=True)
col3.replace(3, "Disagree", inplace=True)
col3.replace(4, "Agree", inplace=True)
col3.replace(5, "Agree", inplace=True)
col3.replace(6, "Agree", inplace=True)

cols = clean_df.drop(['Q3', 'Q17', 'Q76', 'Q84', 'DEMO4','WRKSTRS4', 'AGE', 'JSSPROMO', 'JSSOP', 'JSSSUPER', 'ILEMP5', 0], axis=1)

cols.replace(1, "Strongly Disagree", inplace=True)
cols.replace(2, "Moderately Disagree", inplace=True)
cols.replace(3, "Slightly Disagree", inplace=True)
cols.replace(4, "Slightly Agree", inplace=True)
cols.replace(5, "Moderately Agree", inplace=True)
cols.replace(6, "Strongly Agree", inplace=True)
    
clean_df=pd.concat([col1, col2, col3, cols, clean_df[[0, 'Q3', 'Q76', 'WRKSTRS4', 'AGE', 'JSSPROMO', 'JSSOP', 'JSSSUPER', 'ILEMP5']]], axis=1)

clean_df
```

##### Resetting index and renaming columns

```python
# resetting index
clean_df=clean_df.reset_index(drop=True)
clean_df
```

```python
# Renaming columns
clean_df.set_axis(['HseCare', 'Edu', 'ChgEmp', 'GenDis', 'EthDis', 'MissFamAct', 'PaidFair', 'EmpBft', 'EnjCo', 'InfByOth', 'YrsCurEmp', 'NumChild', 'Stress', 'Age', 'Promo', 'OpProc', 'Supvr', 'IntLvEmp'], axis=1, inplace=True)
clean_df
```
